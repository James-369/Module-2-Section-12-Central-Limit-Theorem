```python
# Import Necessary Libraries 
import numpy as np
from collections import Counter 
import matplotlib.pyplot as plt 
import itertools 
```

```python
pumpkin = ['A', 'B', 'C', 'D', 'E', 'F']
weights = [19, 14, 15, 9, 10, 17]
# Combine both list to create a dictionary 
pumpkin_dict = {}
for i in range(len(pumpkin)):
    pumpkin_dict[pumpkin[i]] = weights[i]
print(pumpkin_dict)
```

```python
# Calculate the population mean from pumpkin_dict 
def calculate_mu(x):
    # Using the formula for mu 
    d = float(sum(x.values())) / len(x)
    return d 
mu = calculate_mu(pumpkin_dict)
mu
```

```python
# Plot a graph illustrating the weights of the pumpkins and highlight the mean weight 
plt.bar(range(len(pumpkin_dict)), pumpkin_dict.values(), align = 'center')
plt.xticks(range(len(pumpkin_dict)), list(pumpkin_dict.keys()))
plt.axhline(y = mu, color = 'r', linestyle = '_',
            label = "Mean Weight:" + str(mu))
plt.legend()
plt.show()
```

```python
def sample_mean(sample_size, data):
    n = sample_size
    # Calcuate the mean population
    mu = calculate_mu(data)
    combs = list(itertools.combinations(data, mu))
    print("Using" , n , "samples with the population size, we can see ", len(combs) , "possible combinations")
    # Calculate the mean weight (xbar) for all combinations (samples) using the given data 
    x_bar_list = []
    # Calculate the sample mean for all combinations 
    for i in range(len(combs)):
        summer = 0
        for j in range(n):
            key_combs = [i][j]
            val = data[str(combs)[i][j]]
            summer += val 
        x_bar = summer / n
        x_bar_list.append(x_bar)
    print("The mean of all sample means mu_x_hat is:", np.mean(x_bar_list))
    return combs, x_bar_list 
n = 2 
# Sample Size 
combs, means = sample_mean(n, pumpkin_dict)
# Print the sample combinations of their means 
for c in range(len(combs)):
    print(c + 1, combs[c], means[c])
    
            
    
```
